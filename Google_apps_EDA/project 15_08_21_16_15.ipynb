## **Data Inspection, Cleaning, and Merging**

The objective of the data inspection and cleaning process is to merge the Google Play app data with the Google Play users' reviews data. 

This process will be divided into the following phases:

* Phase 1: Inspect and clean Google Play app data by handling null values, checking and removing duplicate rows if necessary, excluding unnecessary columns, and modifying column names and values if necessary for ease of analysis.

* Phase 2: Inspect and clean Google Play users' reviews data by handling null values, checking and removing duplicate rows if necessary, excluding unnecessary columns, filtering out unnecessary data, preparing the necessary data for the merging process, modifying column names and values if necessary for ease of analysis.

* Phase 3: Merge the Google Play app data with the Google Play users' reviews data using a common key.  Inspect and clean the combined data.

### **Cleaning the Google Play App Data**

The data frame looks a bit better after replacing the null values.

The null values replacement has to be done by columns since their values' data types are not the same.

Missing numeric values and non-numeric values are replaced by 0 and 'unknown' respectively for ease filtering during the analysis phase.

The next step is to investigate columns that has categorical versions of themselves.

```python
# Using value_counts method to understand the values of columns that have categorical version of themselves
display(googleplay_df['Category'].value_counts())
display(googleplay_df['Category (categorical)'].value_counts())

display(googleplay_df['Type'].value_counts())
display(googleplay_df['Type (categorical)'].value_counts())

display(googleplay_df['Genres'].value_counts())
display(googleplay_df['Genres (categorical)'].value_counts())

display(googleplay_df['Android Ver'].value_counts())
display(googleplay_df['Android Ver (categorical)'].value_counts())

display(googleplay_df['Current Ver'].value_counts())
display(googleplay_df['Current Ver (categorical)'].value_counts())

display(googleplay_df['Last Updated (Year/Month)'].value_counts())
display(googleplay_df['Last Updated (Year/Month) (categorical)'].value_counts())
```

```python
# List of column names that are to be dropped
dropped_columns = ['Last Updated', 'Last Updated (Year/Month) (categorical)']

# Drop columns unnecessary for the analysis
googleplay_df = googleplay_df.drop(dropped_columns, axis=1)
```

```python
modified_col_names = []
col_name = '' 

# Looping through the list of column names and modify the names accordingly
for item in list(googleplay_df.columns):
    if '(categorical)' in item:
        col_name = item.replace(' (categorical)', '')
        col_name = col_name.replace(' ', '_')
        col_name = col_name.lower()
        modified_col_names.append(col_name)
    elif '(Year)' in item:
        col_name = item.replace(' (Year)', '')
        col_name = col_name.replace(' ', '_')
        col_name = col_name.lower()
        modified_col_names.append(col_name)
    else:
        col_name = item.replace(' ', '_')
        col_name = col_name.lower()
        modified_col_names.append(col_name)

# Re-assign the new column names        
googleplay_df.columns = modified_col_names

# Rename the size column name
googleplay_df.rename(columns={'size': 'size_MB',
                              'no_reviews_count' : 'invalid_reviews_count',
                              'reviews_present_count' : 'valid_reviews_count'}, inplace=True)
```

### **Cleaning the Google Play App Users' Reviews Data**

```python
# Drop duplicate rows and check if it has been done properly
googleplay_reviews_df.drop_duplicates(subset=None, keep='first', inplace=True)
display(googleplay_reviews_df[googleplay_reviews_df.duplicated()==True])
display(googleplay_reviews_df.shape)
display(googleplay_reviews_df.info())
```

The duplicated rows are dropped since they will not be useful for analysis purposes.

```python
#List of column names to be included in the dataframe
columns_to_keep = [ 'App',
                    'sentiment_polarity_score', 
                    'sentiment_polarity', 
                    'sentiment_polarity_summarised',
                    'sentiment_subjectivity_score', 
                    'sentiment_subjectivity', 
                    'sentiment_subjectivity_summarised',
                    'original_Sentiment',
                    'original_Sentiment_Polarity',
                    'original_Sentiment_Subjectivity'
                    ]

# Filter the dataframe to only include columns with names stored in the list defined above
# Check if the filtering has been done properly
googleplay_reviews_df = googleplay_reviews_df[columns_to_keep]
display(googleplay_reviews_df.shape)
display(googleplay_reviews_df.info())
```

The values of columns other than the ones included in columns_to_keep list are details about comments stored in the Translated_Review column.

Since the focus of the analysis will not be on comments about the apps, the columns other than the ones included in the columns_to_keep list will be excluded.

```python
# Handling missing values in numerical and non-numerical columns
googleplay_reviews_df['sentiment_polarity_score'] = googleplay_reviews_df['sentiment_polarity_score'].fillna(0)
googleplay_reviews_df['sentiment_polarity'] = googleplay_reviews_df['sentiment_polarity'].fillna('unknown')
googleplay_reviews_df['sentiment_polarity_summarised'] = googleplay_reviews_df['sentiment_polarity_summarised'].fillna('unknown')
googleplay_reviews_df['sentiment_subjectivity_score'] = googleplay_reviews_df['sentiment_subjectivity_score'].fillna(0)
googleplay_reviews_df['sentiment_subjectivity'] = googleplay_reviews_df['sentiment_subjectivity'].fillna('unknown')
googleplay_reviews_df['sentiment_subjectivity_summarised'] = googleplay_reviews_df['sentiment_subjectivity_summarised'].fillna('unknown')
googleplay_reviews_df['original_Sentiment'] = googleplay_reviews_df['original_Sentiment'].fillna('unknown')
googleplay_reviews_df['original_Sentiment_Polarity'] = googleplay_reviews_df['original_Sentiment_Polarity'].fillna(0)
googleplay_reviews_df['original_Sentiment_Subjectivity'] = googleplay_reviews_df['original_Sentiment_Subjectivity'].fillna(0)

# Check to see if the missing values have all been replaced
display(googleplay_reviews_df.shape)
display(googleplay_reviews_df.info())
```

Missing numerical and non-numerical values will be replaced with zeros and 'unknown' respectively for ease of filtering.

```python
# Further filtering to exclude cells with 0 and unknown values
condition1 = googleplay_reviews_df['sentiment_polarity_score'] != 0
condition2 = googleplay_reviews_df['sentiment_polarity'] != 'unknown'
condition3 = googleplay_reviews_df['sentiment_polarity_summarised'] != 'unknown'

googleplay_reviews_df = googleplay_reviews_df[condition1 & condition2 & condition3]

# Check to see if the filtering has been done properly
display(googleplay_reviews_df.shape)
display(googleplay_reviews_df.info())
```

It follows that apps with no polarity data will not have subjectivity data.

These kind of apps can be excluded from the analysis.

```python
# Using value_counts method to understand the sentiment polarity unique values
display(googleplay_reviews_df['sentiment_polarity_score'].value_counts())
display(googleplay_reviews_df['sentiment_polarity'].value_counts())
display(googleplay_reviews_df['sentiment_polarity_summarised'].value_counts())
```

The sentiment_polarity_summarised column's values represent summaries of the sentiment_polarity column. For simplicity during the analysis phase, the sentiment_polarity_summarised column will be used and the sentiment_polarity column will be excluded.

The sentiment_polarity_score column values represent the score range of the sentiment polarity and will be useful for relationship analysis.    



```python
# Create a new data frame with filtered values from the reviews data frame
sentiment_polarity_info = ['App','sentiment_polarity_summarised']
sentiment_polarity_df = googleplay_reviews_df[sentiment_polarity_info]

# Sort and aggregate the sentiment polarity values
sentiment_polarity_df = sentiment_polarity_df.pivot_table(
    index = 'App',
    columns = 'sentiment_polarity_summarised',
    aggfunc = 'size'
)

# Create a new data frame with filtered values from the reviews data frame
sentiment_polarity_info = ['App', 'sentiment_polarity_score']
sentiment_score_df = googleplay_reviews_df[sentiment_polarity_info]

# Sort and aggregate the sentiment polarity score values
sentiment_score_df = sentiment_score_df.groupby('App').mean()

# Merge the sentiment_polarity dataframe with the score dataframe
sentiment_polarity_df = sentiment_polarity_df.merge(sentiment_score_df, on='App')

# Check to see if the sorting and aggregation have been done properly
display(sentiment_polarity_df.shape)
display(sentiment_polarity_df.info())
display(sentiment_polarity_df.head())
```

Dataframe with the sentiment polarity values' distributions for each app is generated in preparation to merge the users' reviews data with the app data.

However, there are still some missing values to be dealt with and an index column to be resetted.

**Question 9:** What is the average rating for each of the top 5 most popular app category and what are the majority of the sentiment polarity on the rating?

```python
condition1 = googleplay_with_sentiment_df['negative_sentiment'] != 0
condition2 = googleplay_with_sentiment_df['neutral_sentiment'] != 0
condition3 = googleplay_with_sentiment_df['positive_sentiment'] != 0

filtered = googleplay_with_sentiment_df[condition1 | condition2 | condition3]
columns_needed = ['category', 'installs', 'rating', 'negative_sentiment', 'neutral_sentiment', 'positive_sentiment']
filtered = filtered[columns_needed]

display(filtered.head())
display(filtered.info())
```

```python
avg_rating = round(filtered.groupby('category').mean()['rating'], 1)

rating_pop_cat = filtered.groupby('category').sum()

rating_pop_cat['rating'] = avg_rating

rating_pop_cat['neg_prop(%)'] = rating_pop_cat['negative_sentiment']/(rating_pop_cat.sum(axis=1) - rating_pop_cat['rating'] - rating_pop_cat['installs'])
rating_pop_cat['neut_prop(%)'] = rating_pop_cat['neutral_sentiment']/(rating_pop_cat.sum(axis=1) - rating_pop_cat['rating'] - rating_pop_cat['installs'])
rating_pop_cat['pos_prop(%)'] = rating_pop_cat['positive_sentiment']/(rating_pop_cat.sum(axis=1) - rating_pop_cat['rating'] - rating_pop_cat['installs'])

rating_pop_cat['neg_prop(%)'] = round(rating_pop_cat['neg_prop(%)']*100, 2)
rating_pop_cat['neut_prop(%)'] = round(rating_pop_cat['neut_prop(%)']*100, 2)
rating_pop_cat['pos_prop(%)'] = round(rating_pop_cat['pos_prop(%)']*100, 2)

rating_pop_cat.sort_values(by='installs', ascending=False, inplace=True)

rating_pop_cat.reset_index(inplace=True)

columns_needed = ['category', 'rating', 'neg_prop(%)', 'neut_prop(%)', 'pos_prop(%)']

proportion = rating_pop_cat[columns_needed]

proportion.set_index(keys=['category','rating'], inplace=True)

top_5 = proportion[0:5]

display(top_5)

top_5.plot(
    kind = 'bar',
    figsize = [10,8],
    stacked = True,
    color = ['cornflowerblue','orchid','gold'],
    fontsize = 14,
    grid = False,
)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title(label= 'Sentiment Polarity Proportion by App Category(%)',fontsize=16)
plt.xlabel('App category with average rating',fontsize=14, labelpad=20)
plt.ylabel('Sentiment polarity proportion(%)',fontsize=14, labelpad=20)

plt.tight_layout()
plt.show()
```

**Answer 9:** The average rating for game category is 4.4, communication is 4.3, tools is 4.3, social is 4.2, and news & magazines is 4.2.  Majority of the sentiment polarity for each of the app category's average rating is positive.

## **Conclusions**

The findings based on the analysis conducted on the Google Play app and user review data are as follows.

The family category has the highest number of apps available and, yet, the game category is more popular. Its popularity can be attributed to the fact that people feel happier, more relaxed, more interested, more focused, and more engaged when playing mobile games than using social apps (Widener 2017).

Free apps have a much bigger proportion than paid apps because free apps attract more targeted users which increase visibility and downloads. They are not as complex as the paid ones and, therefore, less expensive.  Furthermore, they provide more future monetization possibilities and they can help build brand image and name in the online market (Mistry 2021).

Users have much bigger tendencies to download free apps than paid apps because free, ad supported apps can provide users opportunities to see mobile ads that are relevant to them (Sterling 2014). 

Given the facts that content rating is used by consumers, developers, marketers, and advertisers to improve ad targeting by conveying targeted messages to targeted audiences (*App Content Rating Statistics 2021 2021*), Google’s main revenue comes from online advertising, and its platform, AdMob, is utilized by mobile app makers to sell ad space within their apps (Graham and Elias 2021), it makes sense that the majority of the app content rating is everyone since apps with everyone content ratings are accessible to all age group and allows advertisements to be targeted to much wider audiences.

Given that the app dataset accounts for a very small portion of the total number of apps currently available at Google Play, the total paid app revenue indicated at 192.89 million dollars might not reflect the actual total revenue generated by paid apps and family category with everyone 10+ content rating, game category with mature 17+ content rating, and photography category with everyone content rating might not be highest revenue generators.  

However, it is still reasonable to find out how free apps generate revenues when users are not charged for their downloads. The answer to this question is that free apps can generate revenues by using strategies such as advertising, subscriptions, selling merchandise, in-app purchases, sponsorship, referral marketing, collecting and selling data, freemium upsell, physical purchases, transaction fees, and crowdfunding (Mistry 2021).

The user review dataset also accounts for a small portion of the total number of apps currently available at Google Play.  In this case, the subjectivity score seems to be highly concentrated within the range of 0.4 and 0.6 (which indicates neither subjective nor objective). The actual score concentration might not be within this range.  This finding should be revisited when the review data on all the apps listed in the app dataset are included.  

Each of the top 5 app category with the average rating between 4.2 to 4.4 seems to have mostly positive sentiments.  However, the data analysis that resulted in this finding was also done on a small portion of the total number of Google Play apps.  The actual positive sentiment proportion for each of the app category might not be the same. Should more app data be made available, this finding should also be revisited.
